```python
import pandas as pd
import numpy as np
from itertools import product
```

```python
# fusion set covid percentage = 0.139
# ehl_edited set covid percentage = 0.085
```

```python
models = np.array(['/models/COVIDNet-CXR-Gisli'])

datasets = np.array(['fusion_0.8'])

classweights_normal = np.array([1.])
classweights_pneumonia = np.array([5.0])
classweights_covid19 = np.array([3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0])

covid_perc = np.array([0.139])

learning_rates = np.array([1e-5])
batch_sizes = np.array([32])
```

```python
df = pd.DataFrame(list(product(models,
                               datasets,
                               classweights_normal,
                               classweights_pneumonia,
                               classweights_covid19,
                               covid_perc,
                               learning_rates,
                               batch_sizes)), columns=['model', 
                                                       'labels', 
                                                       'classweights_normal', 
                                                       'classweights_pneumonia', 
                                                       'classweights_covid19', 
                                                       'covid_percent', 
                                                       'learning_rate', 
                                                       'batch_size'])

df
```

```python
df.to_csv('training_joblist.csv', index=False)
```
