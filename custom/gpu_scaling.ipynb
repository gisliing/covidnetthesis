```python
plt.figure(figsize=(10,5))

plt.title('COVIDx Inference Scale-up (480x480)')

plt.plot(np.arange(1,11,1), covidx_480)

plt.xlabel('GPUs')
plt.xticks(np.arange(1,11,1))

plt.ylabel('Seconds')
plt.yticks(np.arange(0, 8500, 500))

plt.grid()
plt.show()
```
